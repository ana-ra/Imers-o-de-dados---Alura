Como observamos anteriormente apenas 1866 experimentos tem a classe com_controle para tentar descobrir o motivo vamos tomar como ponto de partida uma analise da coluna 'droga' da base de dados. 

Primeiro vamos contar a frequencia de cada droga com o groupby para ver a distribuicao dos experimentos em cada droga. 

Double-click to edit

````python
droga_group_count_asc = dados.groupby('droga')['id'].count().reset_index(name='Count').sort_values(by='Count',ascending=True)
droga_group_count = dados.groupby('droga')['id'].count().reset_index(name='Count').sort_values(by='Count',ascending=False)
droga_group_count.head()
````

Podemos observar que  no histograma gerado com as 30 primeiras amostras de drogas. Entretando apenas uma pequena por√ß√£o das drogas tem um n√∫mero relativament grande de experimentos. Podemos tamb√©m observar na Fun√ß√£o Distribui√ß√£o Acumulada (CDF) que essas amostras dominantes comp√µem mais ou menos cerca de 20% dos experimentos e uma delas se destaca por compor quase 8% dos experimentos.

````python
fig, axs = plt.subplots(1,2, figsize=(15,6))
axs[0].bar(droga_group_count[:30]['droga'], droga_group_count[:30]['Count']/dados.shape[0], color='green')
axs[0].set_xticks([])
axs[0].set_title("Histograma")
axs[0].set_ylabel("Probabilidade")
axs[1].stem(droga_group_count_asc['Count'].cumsum()/sum(droga_group_count_asc['Count']))
axs[1].grid(True)
axs[1].set_title("Fun√ß√£o de distribui√ß√£o cumulativa")
axs[1].set_ylabel("Probabilidade acumulada")
````

Ap√≥s analisar drogas podemos aferir a frequ√™ncia de ocorr√™ncia de um tratamento com uma droga. J√° √© poss√≠vel ver que um grande n√∫mero de drogas n√£o possui nenhum   grupo de controle.
