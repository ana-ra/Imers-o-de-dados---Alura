Para usarmos uma melhor nomeclatura para representar os compostos, vamos renomear a coluna droga com o nome "Composto". Para isso utilizaremos a função 'map' que existe na biblioteca pandas, essa função trabalha na mudança de um nome. É importante destacar o uso de um parêmetro 'inplace = True', esse parâmetro trabalha com os dados modificando seu locanl no dataframe, atualizando ele. Caso não seja declarado esse parâmetro recebe 'inplace = False' e o retorno será uma cópia do objeto. 

````python
mapa ={'droga':'composto'}
dados.rename(columns=mapa, inplace= True)
````

Para uma melhor vizualização dos dados vamos selecionar apenas os compostos mais frequentes. Vamos trablhar com os 6 mais presentes. Para isso vamos utilizar outro recurso da biblioteca pandas, a funçao 'value-counts', vamos declarar o 'index[0:6]'. Essa parte faz com que a função se atenha somente à contagem dos maiores valores e apresente como resultado apenas o index do intervalo determinado. Ou seja o nome dos 6 maiores valores. 

````python
codcompostos = dados['composto']
````

Para filtrar os dados vamos usar uma função chamada 'query", essa função se assemelha ao SQL. A estrutura dela é bem simplificada, precisando somente definir o drataframa, chamar a funlçao e passar commo parâmetro a condição que deve ser filtrada no nosso conjunto de dados. 

Para o projeto vamos realizar um filtro selecionando somente as linhas em que o composto esteja dentro da lista codcompostos (os 6 compostos mais testados no experimento). 

Como parâmetro para nossa função 'query' vamos passar uma string contendo a lógica para realização da seleção dos dados. O que queremos é o seguinte: o 'query' precisa retornar todas as linhas contendo os 6 compostos. Sendo assim a string será: compostos in @codcompostos. 

Usamos composto porque essa é a coluna a ser verificada no dataframe e codcompostos para ser nossa lista com os 6 compostos, o detalhe aqui é que o @ é necessário para informar o query que codcomposto é uma váriável que já definida fora da função. 

````python
dados.query('composto in @codcompostos')
````

Agora para observação dos gráficos vamos utilizar uma outra biblioteca. A seaborn. 

````python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10,8))
# figsize é um parametro da função figure, ela por sua vez pode receber vários parametros. 
#Esse parametro no caso é uma tupla que representa o eixo x e y e os tamanhos desejados. 
ax = sns.countplot(x ='composto', data=dados.query('composto in @codcompostos'), palette='cubehelix', order=codcompostos)
ax.set_title('6 compostos mais frequentes')
````

Uma estratégia mais direta em que não precisaremos escrever a lista gene a gene, é o uso da função loc[] do Pandas. Como argumentos, passamos primeiramente o :, os dois pontos faz com que o loc[]retorne todos os elementos de uma determinada coluna, isso é importante quando não sabemos qual a quantidade de linhas de um dataframe. E, o segundo elemento, passamos as colunas que são de nosso interesse. No caso, queremos que a função nos retorne todos os elementos das colunas g-0até g-771. E, por fim, podemos declarar a nossa função de interesse a partir deste filtro realizado nos dados, o describe().



````python
dados.loc[:,'g-0':'g-771'].describe()
````

````python
dados.loc[:,'g-0':'g-771'].describe().T
````

````python
plt.figure(figsize=(8,6))
dados.loc[:,'g-0':'g-771'].describe().T['mean'].hist(bins=40, color='green')
````

A seguir, reproduzimos o código acima apenas alterando o parâmetro estatístico a ser analisado (mínimo e máximo, por exemplo). 
E assim, podemos perceber as nuances de cada métrica.

````python
plt.figure(figsize=(8,6))
dados.loc[:,'g-0':'g-771'].describe().T['min'].hist(bins=40, color='green')
````

````python
plt.figure(figsize=(8,6))
dados.loc[:,'g-0':'g-771'].describe().T['max'].hist(bins=30, color='green')
````

Um outro tipo de gráfico super interessante e útil é o boxplot. Para visualizá-lo, vamos usar a função boxplotdo Seaborn e, como argumentos dessa função vamos passar um x, onde x = coluna que será plotada neste eixo e a base de dados data = conjunto de dados. O boxplot apresenta uma caixa no meio onde podemos identificar a mediana (linha no meio da caixa que é o ponto onde metade dos dados estão na direita e a outra metade para a esquerda), os outliers (pontos acima ou abaixo do eixo principal do gráfico que representam valores discrepantes para mais ou para menos), a maior concentração dos dados (caixa principal que representa onde está a mior parte dos dados - primeiro quartil (25%) e terceiro quartil (75%)) e os máximos e mínimos desconsiderando os outliers (linhas laterais à caixa principal). O boxplot é uma importante ferramenta na visualização de dados porque em apenas um gráfico, podemos identificar várias métricas estatísticas.

````python
plt.figure(figsize=(10,8))
sns.boxplot(x='g-0', data=dados,palette='cubehelix' )
````

Podemos também, além de definir apenas os dados que irão no eixo x, definimos os dados para o outro eixo, atribuindo um valor para o parâmetro y (y = variável que vai ser plotada neste eixo). Como podemos perceber, no boxplot que representa o tratamento = com_droga apresenta muitos outliers e isso gera uma discussão bastante interessante pois, do ponto de vista biológico a investigação desses pontos é importante mas, dependendo da área que estamos trabalhando, esse ponto pode apresentar outras soluções. Dito isso, é importante para um cientista de dados não só entender e manipular a base de dados mas também saber acerca do negócio que estamos tratando.

````python
plt.figure(figsize=(10,8))
sns.boxplot(y='g-0', x='tratamento' , data=dados, palette='cubehelix')
plt.title('Relação do tratamento com controle e com droga em relação ao gene0', pad=20)
plt.xlabel('Uso de tratamento')
plt.ylabel('Gene 0')
````

````python
plt.figure(figsize=(10,8))
gra = sns.boxplot(y='g-0', x='tempo' , data=dados, palette='cubehelix')
gra.set_title('Relação do tempo com o gene 0', pad=20)
plt.xlabel('Tempo de incubação')
plt.ylabel('Gene 0')
plt.grid(True)


````

````python
plt.figure(figsize=(10,8))
sns.boxplot(y='g-0', x='dose' , data=dados,palette='cubehelix', order=dados['composto'] )
plt.title('Relação das doses com o  gene 0', pad=10)
plt.xlabel('Dose utilizada')
plt.ylabel('Gene 0')
plt.grid(True)
````

````python
plt.figure(figsize=(10,8))
sns.boxplot(y='g-0', x ='composto', data=dados.query('composto in @codcompostos'), palette='cubehelix')
plt.title('6 pricípais compostos utilizados', pad=10)
plt.xlabel('Compostos')
plt.ylabel('Gene 0')
plt.grid(True)
````
