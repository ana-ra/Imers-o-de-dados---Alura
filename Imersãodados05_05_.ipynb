Para usarmos uma melhor nomeclatura para representar os compostos, vamos renomear a coluna droga com o nome "Composto". Para isso utilizaremos a fun√ß√£o 'map' que existe na biblioteca pandas, essa fun√ß√£o trabalha na mudan√ßa de um nome. √â importante destacar o uso de um par√™metro 'inplace = True', esse par√¢metro trabalha com os dados modificando seu locanl no dataframe, atualizando ele. Caso n√£o seja declarado esse par√¢metro recebe 'inplace = False' e o retorno ser√° uma c√≥pia do objeto.

Para filtrar os dados vamos usar uma fun√ß√£o chamada 'query", essa fun√ß√£o se assemelha ao SQL. A estrutura dela √© bem simplificada, precisando somente definir o drataframa, chamar a funl√ßao e passar commo par√¢metro a condi√ß√£o que deve ser filtrada no nosso conjunto de dados. 

Para o projeto vamos realizar um filtro selecionando somente as linhas em que o composto esteja dentro da lista codcompostos (os 6 compostos mais testados no experimento). 

Como par√¢metro para nossa fun√ß√£o 'query' vamos passar uma string contendo a l√≥gica para realiza√ß√£o da sele√ß√£o dos dados. O que queremos √© o seguinte: o 'query' precisa retornar todas as linhas contendo os 6 compostos. Sendo assim a string ser√°: compostos in @codcompostos. 

Usamos composto porque essa √© a coluna a ser verificada no dataframe e codcompostos para ser nossa lista com os 6 compostos, o detalhe aqui √© que o @ √© necess√°rio para informar o query que codcomposto √© uma v√°ri√°vel que j√° definida fora da fun√ß√£o. 

````python
dados.query('composto in @codcompostos')
````

Agora para observa√ß√£o dos gr√°ficos vamos utilizar uma outra biblioteca. A seaborn. 

````python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set()
plt.figure(figsize=(10,8))
# figsize √© um parametro da fun√ß√£o figure, ela por sua vez pode receber v√°rios parametros. 
#Esse parametro no caso √© uma tupla que representa o eixo x e y e os tamanhos desejados. 
ax = sns.countplot(x ='composto', data=dados.query('composto in @codcompostos'), palette='cubehelix', order=codcompostos)
ax.set_title('6 compostos mais frequentes')
````

Uma estrat√©gia mais direta em que n√£o precisaremos escrever a lista gene a gene, √© o uso da fun√ß√£o loc[] do Pandas. Como argumentos, passamos primeiramente o :, os dois pontos faz com que o loc[]retorne todos os elementos de uma determinada coluna, isso √© importante quando n√£o sabemos qual a quantidade de linhas de um dataframe. E, o segundo elemento, passamos as colunas que s√£o de nosso interesse. No caso, queremos que a fun√ß√£o nos retorne todos os elementos das colunas g-0at√© g-771. E, por fim, podemos declarar a nossa fun√ß√£o de interesse a partir deste filtro realizado nos dados, o describe().



````python
dados.loc[:,'g-0':'g-771'].describe()
````

````python
dados.loc[:,'g-0':'g-771'].describe().T
````

````python
plt.figure(figsize=(8,6))
dados.loc[:,'g-0':'g-771'].describe().T['mean'].hist(bins=40, color='green')
````

A seguir, reproduzimos o c√≥digo acima apenas alterando o par√¢metro estat√≠stico a ser analisado (m√≠nimo e m√°ximo, por exemplo). 
E assim, podemos perceber as nuances de cada m√©trica.

````python
plt.figure(figsize=(8,6))
dados.loc[:,'g-0':'g-771'].describe().T['min'].hist(bins=40, color='green')
````

````python
plt.figure(figsize=(8,6))
dados.loc[:,'g-0':'g-771'].describe().T['max'].hist(bins=30, color='green')
````

√â muito interessante que a gente replique a an√°lise desenvolvida para os genes (g), nos tipos celulares (c).
````python
plt.figure(figsize=(8,6))
dados.loc[:,'c-0':'c-99'].describe().T['mean'].hist(bins=50, color='green')
````

Um outro tipo de gr√°fico super interessante e √∫til √© o boxplot. Para visualiz√°-lo, vamos usar a fun√ß√£o boxplotdo Seaborn e, como argumentos dessa fun√ß√£o vamos passar um x, onde x = coluna que ser√° plotada neste eixo e a base de dados data = conjunto de dados. O boxplot apresenta uma caixa no meio onde podemos identificar a mediana (linha no meio da caixa que √© o ponto onde metade dos dados est√£o na direita e a outra metade para a esquerda), os outliers (pontos acima ou abaixo do eixo principal do gr√°fico que representam valores discrepantes para mais ou para menos), a maior concentra√ß√£o dos dados (caixa principal que representa onde est√° a mior parte dos dados - primeiro quartil (25%) e terceiro quartil (75%)) e os m√°ximos e m√≠nimos desconsiderando os outliers (linhas laterais √† caixa principal). O boxplot √© uma importante ferramenta na visualiza√ß√£o de dados porque em apenas um gr√°fico, podemos identificar v√°rias m√©tricas estat√≠sticas.

````python
plt.figure(figsize=(10,8))
sns.boxplot(x='g-0', data=dados,palette='cubehelix' )
````

Podemos tamb√©m, al√©m de definir apenas os dados que ir√£o no eixo x, definimos os dados para o outro eixo, atribuindo um valor para o par√¢metro y (y = vari√°vel que vai ser plotada neste eixo). Como podemos perceber, no boxplot que representa o tratamento = com_droga apresenta muitos outliers e isso gera uma discuss√£o bastante interessante pois, do ponto de vista biol√≥gico a investiga√ß√£o desses pontos √© importante mas, dependendo da √°rea que estamos trabalhando, esse ponto pode apresentar outras solu√ß√µes. Dito isso, √© importante para um cientista de dados n√£o s√≥ entender e manipular a base de dados mas tamb√©m saber acerca do neg√≥cio que estamos tratando.

````python
plt.figure(figsize=(10,8))
sns.boxplot(y='g-0', x='tratamento' , data=dados, palette='cubehelix')
plt.title('Rela√ß√£o do tratamento com controle e com droga em rela√ß√£o ao gene0', pad=20)
plt.xlabel('Uso de tratamento')
plt.ylabel('Gene 0')
````

````python
plt.figure(figsize=(10,8))
gra = sns.boxplot(y='g-0', x='tempo' , data=dados, palette='cubehelix')
gra.set_title('Rela√ß√£o do tempo com o gene 0', pad=20)
plt.xlabel('Tempo de incuba√ß√£o')
plt.ylabel('Gene 0')
plt.grid(True)


````

````python
plt.figure(figsize=(10,8))
sns.boxplot(y='g-0', x='dose' , data=dados,palette='cubehelix', order=dados['composto'] )
plt.title('Rela√ß√£o das doses com o  gene 0', pad=10)
plt.xlabel('Dose utilizada')
plt.ylabel('Gene 0')
plt.grid(True)
````

````python
plt.figure(figsize=(10,8))
sns.boxplot(y='g-0', x ='composto', data=dados.query('composto in @codcompostos'), palette='cubehelix')
plt.title('6 pric√≠pais compostos utilizados', pad=10)
plt.xlabel('Compostos')
plt.ylabel('Gene 0')
plt.grid(True)
````
